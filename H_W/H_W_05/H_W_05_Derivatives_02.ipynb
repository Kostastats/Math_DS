###### Постройте график трехмерный график функции:

f(X)= x(2)^2 - x(1)^2x(2) + 2x(1)x(2)

А также график линий уровня с уровнями {1,5,10}

```python
def f(x1, x2):
    return np.cos(x2**2 - x1**2*x2 + 2*x1*x2)


x = np.linspace(1, 10, 10, 10)
y = np.linspace(1, 10, 10, 10)

X, Y = np.meshgrid(x, y)
Z = f(X, Y) 

fig = plt.figure()
fig.set_figwidth(10)
fig.set_figheight(10)
ax = plt.axes(projection='3d')
#ax.contour3D(X, Y, Z, levels= 1)
ax.contour(X, Y, Z,  levels = 20)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')


surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=1, aspect=10)
ax.azim = 20
ax.dist = 5
ax.elev = 30
```

##### график линий уровня с уровнями 1

```python
fig, ax = plt.subplots()

ax.contour(Z, levels = 1)

fig.set_figwidth(6)    #  ширина и
fig.set_figheight(6)    #  высота "Figure"

plt.show()
```
