```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set()
from pylab import rcParams 
rcParams['figure.figsize'] = 10, 6
```

```python
some_var = stats.norm(loc= 100, scale= 20)
```

```python
some_var.rvs(10)
```

```python
that_var = np.trunc(some_var.rvs(10))
that_var.mean()
```

```python
that_var.std()
```

```python
sns.histplot(that_var, bins=50, color='Green')
```

```python
that_var50 = np.trunc(some_var.rvs(50))
that_var50.mean()
```

```python
that_var50.std()
```

```python
sns.histplot(that_var50, bins=50, color='Green')
```

```python
that_var250 = np.trunc(some_var.rvs(250))
that_var250.mean()
```

```python
that_var250.std()
```

```python
sns.histplot(that_var250, bins=50, color='Green')
```

###### Видно как при увеличении множества среднее приблежается к центральному значению и вся выборка приобретает форму Гаусса

##### При каком значение n полученное распредление похоже на нормальное?

### Значение n = 1

```python
mList = []
for kk in range(1000):
    data = exp_rv.rvs(size=1)
    mList.append(data.mean())    # заполняем список средними по нашей выборке
    
sns.histplot(x=mList)
```
