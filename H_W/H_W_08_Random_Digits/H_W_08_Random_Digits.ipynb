```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
n_bins = 15
n, p = 1, 0.5
head_tails = np.random.binomial(n, p, size=n_bins)
h=plt.hist(head_tails,bins=n_bins, density=True)
```

```python
head_tails.mean()
```

```python
head_tails.std()
```

```python
import seaborn as sns
sns.distplot(head_tails)
plt.show()
```
